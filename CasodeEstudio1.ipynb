0. Cargar Librerias

```python
import numpy as np
import pandas as pd
from  sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Montar Base de Datos

```python
nxl="/content/0. SolicitantesCrédito(USD)_NB.xlsx"
XDB=pd.read_excel(nxl,sheet_name=1)
XDB=XDB.dropna()

#Se selecionan las Variables
XD=XDB[['Edad', 'Ingresos', 'Egresos', 'Monto (EAD)', 'Plazo', 'Cuota (COP)']]
yd=np.array(XDB[['Estrato']])
```

2. Se crea el Modelo de Clasificación Naive Bayes

```python
mnb=GaussianNB()
mnb.fit(XD,yd)

u=mnb.theta_
print("Las medias de las variables son:",u)
#La categoría de clasificación que tiene la edad más alta es la 0, con una media de 42.627.
#La categoría de clasificación que posee los ingresos más altos es la 5 con una media de 1533.655.
#La categoría que posee los egresos más bajos es la 0 con una media de 229.461.

sigma=np.sqrt(mnb.var_)
print("La desviación estandar de cada una de las variables son:",sigma)

#Datos Predecibles
ymnb=mnb.predict(XD)

#Matriz de Confusión
confusion_matrix(yd,ymnb)
#La categoría que tuvo la mayor cantidad de aciertos es la 2 (925).

#Cantidad de Solicitantes por Estrato
n=mnb.class_count_
print("La cantidad de solicitantes por estrato es",n)
#La categoría que posee mayor cantidad de solicitantes es la 2, con 2419 solicitantes.
#El estrato, que posee la mayor cantidad de solicitantes es el estrato 3. Y el menor es el estrato 6.
#La categoría que posee la menor cantidad de solicitantes es la 5, con 20 participantes.
```

3. Evaluamos el comportamiento del modelo

```python
#Se evalua a que estrato pertenece un individuo con las siguientes características: Edad=34, Ingresos=850 USD,
#Egresos=250USD, Monto=1250, Plazo=24, Cuota=52
XA=np.array([[34,850,250,1250,24,52]])
Vp=mnb.predict(XA)
print("La categoría a la que pertence la persona con estas características es: ",Vp)
#Esta persona pertence al estrato 4.
```

4. Descargar la matriz de confusion a Excel

```python
cm=confusion_matrix(yd,ymnb)
dfcm=pd.DataFrame(cm)
dfcm.to_excel("Reto1.xlsx")

#El estrato que logró mayor cantidad de aciertos frente al modelo Naive Bayes es el estrato 1.
```
